```python
import pandas as pd
import streamlit as st
```

```python
df = pd.read_csv('player_data_final.csv')
similarity_indexes = pd.read_csv('similarity_matrix.csv').to_numpy()
```

```python
def recommend(name,size=20,potential=False):
    index = df[df['Name']==name].index.tolist()[0]
    most_similar_indexes = similarity_indexes[index]
    if potential:
        return df.iloc[most_similar_indexes[1:],:4].where(df.Age<26).sort_values('POT',ascending=False).iloc[:size]
    else:
        return df.iloc[most_similar_indexes[1:size+1],:4]
```

```python
st.title("Player Recommender")
name =st.text_input('Enter Player Name', 'Ronaldo')
potential = st.checkbox('Recommend me only young players (<26 years old)')
if name:
    recommend(name,potential=potential)
```
